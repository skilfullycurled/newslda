```python
outlets = ['bnn', 'wsj', 'cnn']
# lda_path = 'data/lda/gensim/'
lda_path = Path('data/lda/gensim/')
files = [filename.stem for filename in lda_path.iterdir() if (filename.is_file() and (filename.suffix == '.lda'))]
headers = ['outlet', 'level', 'stoptype', 'topic', 'terms']

outlet_dict = create_outlet_dict(files, outlets)

for outlet, files in outlet_dict.items():
    
    rows = []

    for level, filename in enumerate(files):
        
#         print level, filename
        
        outlet_lda = models.LdaModel.load('data/lda/gensim/' + filename + '.lda')
        outlet_topics = outlet_lda.show_topics(formatted=False)

        file_parts = filename.split('_')

        outlet = file_parts[0]
        stoptype = file_parts[-1]

        for topic_no, terms in enumerate(outlet_topics):
    #         ['bnn', 0, 'full', 0, u'the to of election in and for that elections on']
            row = [outlet, level, stoptype, topic_no, ' '.join([term[0] for term in terms[1]])]
            rows.append(row)
            

    topics_df = pd.DataFrame(rows, columns=headers)
    topics_df.to_csv('data/topics/' + outlet + '_topics.csv', index=False)
```

```python
def create_outlet_dict(filenames, outlets):
    
    outlet_files = {}

    for outlet in outlets:

        outlet_files[outlet] = [filename for filename in filenames if outlet in filename]
        
    return outlet_files


# test_files = create_outlet_dict(files, outlets)

# for outlet, files in test_files.items():
#     print outlet, files
```
